# 열 및 행선택
- CRIM, ZN, INDUS 열 선택하기

```python
columns = ["CRIM", "ZN", "INDUS"]
result = df_boston[columns]
result.head()
```

- ZN == 18.0 에 해당하는 행만 추출

```python
result = df_boston[df_boston["ZN"] == 18.0]
result
# .loc 있을 떄, 없을 때의 차이
# loc를 사용하는 것이 코드의 가독성을 높이고, 행과 열을 동시에 선택해야 할 때 더 명확한 방법임.
```

- AGE >= 50 에 해당하는 데이터 추출

```python
result = df_boston[df_boston["AGE"] >= 50].reset_index(drop=True)
result
```

- INDUS가 5.0 이상, B가 394 이상인 데이터 추출

```python
result = df_boston[(df_boston["INDUS"] >= 5.0) & (df_boston['B'] >= 394)].reset_index(drop=True)
# result = df_boston[(조건식1) & (조건식2)...] 조건이 많아졌을 때 사용.
result
```

## iloc와 loc 문법
- iloc : Purely integer-location based indexing for selection by position.
.iloc[] is primarily integer position based (from 0 to length-1 of the axis), but may also be used with a boolean arra

- loc : Access a group of rows and columns by label(s) or a boolean array.
.loc[] is primarily label based, but may also be used with a boolean array.y.

```python
# iloc를 활용한 추출(index, columns)
result = df_boston.iloc[:3, :3]
result
```

```python
# loc를 활용한 추출(index, columns)
result = df_boston.loc[:2, 'CRIM':'INDUS']
result
```

```python
# 조건을 적용한 loc 추출도 가능/ iloc는 조건이 안돼서???였나????? 확인하기
result = df_boston.loc[(df_boston['INDUS'] >= 5.0) & (df_boston['B'] >= 394) # 가독성
                                ,['CRIM', 'ZN', 'INDUS']].reset_index(drop=True)
result.head()
```

```python
temp_df['shifted_v1'] = temp_df['price'].shift(1, fill_value=0).astype(int)
temp_df['diff'] = temp_df['price'] - temp_df['shifted_v1']
temp_df.head()
```

```python
temp_df['shifted_v2'] = temp_df['price'].shift(2, fill_value=0).astype(int)
temp_df['diff2'] = temp_df['price'] - temp_df['shifted_v2']
temp_df.head()
```

### 코로나 관련 변수만 추출
- 결측값이 있는지 확인
- 결측값이 있는 국가 하나 존재.
- tovars 사용

```python
# totvars = ['location','total_cases','total_deaths','total_cases_pm','total_deaths_pm']
totvars_misscnt = covidtotals[totvars].isnull().sum(axis=1)
# totvars_misscnt.value_counts()
covidtotals.loc[totvars_misscnt >= 1, totvars].T
```

```python
covidtotalsonly = covidtotals.loc[:, totvars]
covidtotalsonly['total_cases_q'] = pd.qcut(covidtotalsonly['total_cases'], 
                                           labels = ['매우 낮음', '낮음','중간','높음','매우 높음'],
                                          q=5, precision=0)

covidtotalsonly['total_deaths_q'] = pd.qcut(covidtotalsonly['total_deaths'], 
                                           labels = ['매우 낮음', '낮음','중간','높음','매우 높음'],
                                          q=5, precision=0)
covidtotalsonly.head()
```

```python
pd.crosstab(covidtotalsonly['total_cases_q'], covidtotalsonly['total_deaths_q'])

# 이상치에 대한 정의는 다양하지만, 여기서는 매우낮음-매우낮음 ~ 매우높음-매우높음 이어지는 우하향선
# 우하향선 기준으로 두 칸 이상 떨어진 국가들을 이상치로 정하여 조사를 해보고자 함.
```

```python
# 사례 하나. 낮음 = 높음 경우
covidtotals.loc[(covidtotalsonly.total_cases_q=="낮음") & (covidtotalsonly.total_deaths_q=="높음")]
```

```python
covidtotalsonly['total_cases_i'] = covidtotalsonly['total_cases_q'].map({'매우 낮음': 1, '낮음': 2, '중간': 3, '높음': 4, '매우 높음': 5}).astype(int)
covidtotalsonly['total_deaths_i'] = covidtotalsonly['total_deaths_q'].map({'매우 낮음': 1, '낮음': 2, '중간': 3, '높음': 4, '매우 높음': 5}).astype(int)
```

```python
covidtotalsonly['diff_abs'] = abs(covidtotalsonly['total_cases_i'] - covidtotalsonly['total_deaths_i'])
covidtotalsonly[covidtotalsonly['diff_abs'] >= 2]
```
