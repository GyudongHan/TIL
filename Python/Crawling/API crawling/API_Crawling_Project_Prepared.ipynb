```python
import requests
import json
import pandas as pd
SERVICE_KEY = 'your_key'

data_list = []

for start_idx in range(1, 10001, 1000):
    end_idx = min(start_idx + 999, 10000)
    url = f'http://openapi.seoul.go.kr:8088/{SERVICE_KEY}/json/VwsmTrdarFlpopQq/{start_idx}/{end_idx}/2017'
    content = requests.get(url).json()
    data_list.extend(content.get('VwsmTrdarFlpopQq', {}).get('row', []))

result = pd.DataFrame(data_list)
print(result)

# DataFrame을 CSV 파일로 저장
result.to_csv('생활인구_2017.csv', index=False)
```
